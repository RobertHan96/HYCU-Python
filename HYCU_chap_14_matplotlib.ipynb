# HYCU14강 Matplotlib 패키지 활용 예제

```python
import matplotlib.pyplot as plt
```

## X 축에 대응하는 Y축의 2차원 그래프 생성

```python
x = list(range(1,21))
y= [value**2 for value in x]
plt.plot(x,y)
```

## 패키지의 함수들을 통해 그래프 꾸미기

```python
graph1 = plt.plot(x,y,'ro') #색상과 마커(빨강, 원형)로 그래프 표시
```

```python
graph2 = plt.plot(x,y,'go')
```

```python
plt.xlabel('Value')
plt.ylabel('Squared of X') # x, y 축의 캡션 달기
```

```python
plt.grid() # 눈금선 표시하기
```

```python
graph3 = plt.plot(x,y,'bo')
```
